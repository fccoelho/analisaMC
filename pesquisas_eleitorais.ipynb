```python
def sphinx_query(index_name, query="", facet=None):
    """
    Search Sphinx index using a simple match via SphinxQL
    :param index_name: Name of the index to search on
    :param query: String with the query expression
    :param facet: Attribute name to facet by. Must be a list
    :return: JSON (array of objects)
    """
    try:
        assert index_name in ['mediacloud_articles', 'mediacloud_feeds', 'mediacloud_tweets']
    except AssertionError:
        return json.dumps({"error": "Bad index name: {}".format(index_name)})
    

    # Setup Sphinxsearch SphinxQL connection
    sphinx_conn = pymysql.connect(host='200.20.164.152', port=9306)
    cursor = sphinx_conn.cursor(pymysql.cursors.DictCursor)

    if facet is None:
        cursor.execute("SELECT * from " + index_name + " WHERE MATCH(%(query)s) "
                "LIMIT %(limit)s OPTION max_matches=%(limit)s",
                    {'query': query, 'limit': 100000})
    else:
        cursor.execute("SELECT * from "+index_name+" WHERE MATCH(%s) " + " ".join(["FACET {}".format(f) for f in facet]),
                       (query,))
    results = cursor.fetchall()
    cursor.close()
    
    return json.dumps(results)
```

```python
class MLStripper(HTMLParser.HTMLParser):
    def __init__(self):
        #super(HTMLParser.HTMLParser).__init__(self)
        self.reset()
        self.fed = []
        self.strict = False

    def handle_data(self, d):
        self.fed.append(d)

    def get_data(self):
        return ''.join(self.fed)

def strip_tags(H):
    s = MLStripper()
    s.feed(H)
    return s.get_data()
```

```python
def consulta_sphinx(consulta, data_ini, data_fim):
    res = sphinx_query("mediacloud_articles",'"{}"'.format(consulta))
    res = json.loads(res)
    # retemos apenas as entradas que contém sumários
    res = [d for d in res if dt.datetime.fromtimestamp(d['published'])>= data_ini and 
           dt.datetime.fromtimestamp(d['published']) < data_fim]
    print '{} Artigos retidos com a consulta \"{}\"'.format(len(res), consulta)
    ts = pd.TimeSeries(data=np.ones(len(res)),index=[datetime.datetime.fromtimestamp(d['published']) for d in res])
    ts = ts.resample('w',how="sum",)
    return ts
```

```python
consultas = ['Marina Silva', 'Eduardo Campos', 'Dilma', 'Aecio']
data_ini = datetime.datetime(2014,4,5,0,0,0)
data_fim = datetime.datetime(2014,8,26,23,59,59)
ts_time_span = pd.date_range(start=data_ini, end=data_fim, freq='W')
df_media = pd.DataFrame(index=ts_time_span)
df_media.index.name = 'Datas'
```

```python
for consulta in consultas:
    ts_volumes = consulta_sphinx(consulta, data_ini, data_fim)
    df_media[consulta]=ts_volumes

df_media['Total'] = df_media.sum(axis=1, skipna=True, numeric_only=True)
df_media
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(16,8))
df_media.plot(ax=axes)
```

```python
df_media2 = df_media.copy()
for consulta in consultas:
    df_media2[consulta]=(df_media2[consulta]/df_media2['Total'])
    df_media2[consulta] = pd.Series([round(val, 2) for val in df_media2[consulta]], index=df_media2.index)
    #df_media2[consulta] = pd.Series(["{0:.2f}%".format(val * 100) for val in df_media2[consulta]], index=df_media2.index)
del(df_media2['Total'])
df_media2
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(16,8))
df_media2.plot(ax=axes)
```
